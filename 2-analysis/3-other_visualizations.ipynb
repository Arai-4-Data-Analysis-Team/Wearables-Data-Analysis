```python
# read data (from unlabeled dataset)

df_unlabeled = pd.read_csv('../data/unlabeled/059-COMBINED-2022-11-9-7-4-48-1667995631-v1.4.3+lucina-buzz.csv', index_col=0)
```

```python
# add 'seconds' column

counter = 1
seconds = 1
sec_list = []
for x in range(len(df_unlabeled)):
    if counter <= 25:
        sec_list.append(seconds)
        counter = counter + 1
    else:
        seconds = seconds + 1
        sec_list.append(seconds)
        counter = 2
df_unlabeled['seconds'] = sec_list

# add alpha_r and beta_r difference column

diff_list = []
index = 0
for a in df_unlabeled['alpha_r']:
    diff = a - df_unlabeled.loc[index, 'beta_r']
    diff_list.append(diff)
    index = index + 1
df_unlabeled['diff_alpha_beta'] = diff_list
```

```python
# calculate moving averages for 'acc_x_n' values

acc_x_n = df_unlabeled['acc_x_n'].to_frame()
acc_x_n['sma'] = acc_x_n['acc_x_n'].rolling(125).mean()
acc_x_n['cma'] = acc_x_n['acc_x_n'].expanding().mean()
acc_x_n['ema'] = acc_x_n['acc_x_n'].ewm(span=125).mean()

# calculate moving averages for 'acc_y_n' values

acc_y_n = df_unlabeled['acc_y_n'].to_frame()
acc_y_n['sma'] = acc_y_n['acc_y_n'].rolling(125).mean()
acc_y_n['cma'] = acc_y_n['acc_y_n'].expanding().mean()
acc_y_n['ema'] = acc_y_n['acc_y_n'].ewm(span=125).mean()

# calculate moving averages for 'acc_z_n' values

acc_z_n = df_unlabeled['acc_z_n'].to_frame()
acc_z_n['sma'] = acc_z_n['acc_z_n'].rolling(125).mean()
acc_z_n['cma'] = acc_z_n['acc_z_n'].expanding().mean()
acc_z_n['ema'] = acc_z_n['acc_z_n'].ewm(span=125).mean()
```

```python
# plot of 'acc_x_n' vs moving averages

x = sec_list
y = df_unlabeled['acc_x_n']
y1 = acc_x_n['sma']
y2 = acc_x_n['cma']
y3 = acc_x_n['ema']
plt.figure(figsize=(16, 8))
plt.plot(x, y, label='acc_x_n')
plt.plot(x, y1, label='sma')
plt.plot(x, y2, label='cma')
plt.plot(x, y3, label='ema')
plt.xlabel('seconds')
plt.ylabel('acc_x_n moving average')
plt.legend()
plt.show()
```

```python
# plot of 'acc_y_n' vs moving averages

x = sec_list
y = df_unlabeled['acc_y_n']
y1 = acc_y_n['sma']
y2 = acc_y_n['cma']
y3 = acc_y_n['ema']
plt.figure(figsize=(16, 8))
plt.plot(x, y, label='acc_y_n')
plt.plot(x, y1, label='sma')
plt.plot(x, y2, label='cma')
plt.plot(x, y3, label='ema')
plt.xlabel('seconds')
plt.ylabel('acc_y_n moving average')
plt.legend()
plt.show()
```

```python
# plot of 'acc_z_n' vs moving averages
x = sec_list
y = df_unlabeled['acc_z_n']
y1 = acc_z_n['sma']
y2 = acc_z_n['cma']
y3 = acc_z_n['ema']
plt.figure(figsize=(16, 8))
plt.plot(x, y, label='acc_z_n')
plt.plot(x, y1, label='sma')
plt.plot(x, y2, label='cma')
plt.plot(x, y3, label='ema')
plt.xlabel('seconds')
plt.ylabel('acc_z_n moving average')
plt.legend()
plt.show()
```

```python
# plot of 'alpha_r' and 'beta_r' over time (seconds)

x = sec_list
y1 = df_unlabeled['alpha_r']
y2 = df_unlabeled['beta_r']
plt.figure(figsize=(16, 8))
plt.plot(x, y1, label='alpha_r')
plt.plot(x, y2, label='beta_r')
plt.xlabel('seconds')
plt.ylabel('angle')
plt.legend()
plt.show()
```
