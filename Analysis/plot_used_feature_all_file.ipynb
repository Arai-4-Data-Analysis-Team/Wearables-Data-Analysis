```python
# Import the necessary libraries
import os
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# specify the folder where the CSV files are located
folder = 'path/to/folder/'

# create an empty list to store the dataframes
df_list = []

# loop through all the files in the folder
for file in os.listdir(folder):

    if file.endswith('.csv'):
        df = pd.read_csv(os.path.join(folder, file))
        df_list.append(df)

# concatenate all the dataframes together
final_df = pd.concat(df_list)


# Prepare the variables to plot to choose the appropriate features for the model

df_all_acc_r = final_df.groupby(['label'])['acc_x_r','acc_y_r','acc_z_r'].median()

df_all_acc_n = final_df.groupby(['label'])['acc_x_n','acc_y_n','acc_z_n'].median()

df_all_gyr_r = final_df.groupby(['label'])['gyr_x_r','gyr_y_r','gyr_z_r'].median()

df_all_gyr_n = final_df.groupby(['label'])['gyr_x_n','gyr_y_n','gyr_z_n'].median()

df_all_difference_alpha_beta = final_df.groupby(['label'])['different_alpha_beta'].median()
```

```python
# plot acc R  data in a bar chart

df_all_acc_r.plot.bar(color=('steelblue','crimson','mediumspringgreen'),
title='The average of acceleration x/y/z in belt sensor based on label')
plt.show()
```

```python
# Create acc R scatter plot
sns.scatterplot(x = 'acc_x_r' , y='acc_y_r', hue='label', data= final_df , palette="Set2",)
sns.set(rc={"figure.figsize":(30, 10)})
plt.title("The average of acceleration x/y in belt sensor based on label")
plt.show()
```

```python
# plot acc N  data in a bar chart

df_all_acc_n.plot.bar(color=('steelblue','mediumspringgreen','crimson'),
title='The average of acceleration x/y/z in neck based on label')
plt.show()
```

```python
# Create acc N scatter plot
sns.scatterplot(x = 'acc_x_n' , y='acc_y_n', hue='label', data= final_df , palette="Set2")
sns.set(rc={"figure.figsize":(30, 10)})
plt.title("The average of acceleration x/y in neck based on label")
plt.show()
```

```python
# plot gyr R data in a bar chart

df_all_gyr_r.plot.bar(color=('steelblue','crimson','mediumspringgreen'),
title='The average of gyroscope x/y/z in belt sensor based on label')
plt.show()
```

```python
# Create gyr R scatter plot

sns.scatterplot(x = 'gyr_x_r' , y='gyr_y_r', hue='label', data= final_df , palette="Set2")
sns.set(rc={"figure.figsize":(30, 10)})
plt.title("The average of gyroscope x/y in belt sensor based on label")
plt.show()
```

```python
# plot gyr N data in a bar chart

df_all_gyr_n.plot.bar(color=('steelblue','crimson','mediumspringgreen'),
title='The average of gyroscope x/y/z in neck based on label')
plt.show()
```

```python
# Create gyr N scatter plot

sns.scatterplot(x = 'gyr_x_n' , y='gyr_y_n', hue='label', data= final_df , palette="Set2")
sns.set(rc={"figure.figsize":(30, 10)})
plt.title("The average of gyroscope x/y in neck based on label")
plt.show()
```

```python
# plot the difference between alpha and beta data in a bar chart

df_all_difference_alpha_beta.plot.bar(color=('crimson'),
title='The difference between Alpha R and Beta R')
plt.show()
```
